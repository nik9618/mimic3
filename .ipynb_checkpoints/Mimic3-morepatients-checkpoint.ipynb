```python
def processing(aids):
    try:
        conn = psycopg2.connect("dbname='mimic' user='mimic' host='localhost' password='meladymimic315'")
    except:
        print "I am unable to connect to the database"
    print 'connection established'
    
    DB = []
    ct =0 ;

    for aid in aids:

        # admission
        cur = conn.cursor()
        cur.execute('SELECT * FROM mimiciii.admissions WHERE hadm_id='+str(aid))
        admission = cur.fetchone()
        admitdate = admission[3]
        if(admitdate == None):
            continue;
        print str(ct) + ": "+ str(aid)
        ct+=1
        wholedata = []

        cur = conn.cursor()
        cur.execute('SELECT starttime,itemid,amount,amountuom FROM mimiciii.inputevents_mv WHERE amount>0 and hadm_id = '+str(aid)+' and itemid in ('+input_itemid_list+')')
        inputevents = cur.fetchall()
        for ie in inputevents : 
            if(ie[0] == None):
                print 'ie date 0'
                continue;
            wholedata.append(['ie',(ie[0]-admitdate).total_seconds(),ie])

        cur = conn.cursor()
        cur.execute('SELECT charttime,itemid,value FROM mimiciii.outputevents WHERE hadm_id = '+str(aid)+' and itemid in ('+output_itemid_list+')')
        outputevents = cur.fetchall()
        for oe in outputevents : 
            if(oe[0] == None):
                print 'oe date 0'
                continue;
            wholedata.append(['oe',(oe[0]-admitdate).total_seconds(),oe])

        cur = conn.cursor()
        cur.execute('SELECT charttime,itemid,valuenum,valueuom FROM mimiciii.labevents WHERE hadm_id = '+str(aid)+' and itemid in ('+lab_itemid_list+')')
        labevents = cur.fetchall()
        for le in labevents : 
            if(le[0] == None):
                print 'le date 0'
                continue;
            wholedata.append(['le',(le[0]-admitdate).total_seconds(),le])

        cur = conn.cursor()
        cur.execute('SELECT starttime,formulary_drug_cd,dose_val_rx,dose_unit_rx FROM mimiciii.prescriptions WHERE hadm_id = '+str(aid)+' and formulary_drug_cd in ('+prescript_itemid_list+')')
        presevents = cur.fetchall()
        for pe in presevents: 
            if(pe[0] == None):
                print 'pe date 0'
                continue;
            wholedata.append(['pe',(pe[0]-admitdate).total_seconds(),pe])

        cur = conn.cursor()
        cur.execute('SELECT charttime,itemid,valuenum,valueuom FROM mimiciii.chartevents WHERE hadm_id = '+str(aid)+' and itemid in ('+chart_itemid_list+')')
        chartevents = cur.fetchall()
        for ce in chartevents: 
            if(ce[0] == None):
                print 'ce date 0'
                continue;
            wholedata.append(['ce',(ce[0]-admitdate).total_seconds(),ce])

        for mid in microbio_itemid:
            m = mid.split('-')
            cur = conn.cursor()
            sql = 'SELECT charttime,concat(spec_itemid,\'-\',org_itemid,\'-\',ab_itemid),dilution_text,\'uom\' FROM mimiciii.microbiologyevents WHERE hadm_id = '+str(aid)+' and spec_itemid ='+m[0]+' and org_itemid = '+ m[1]
            if(m[2] == ''):
                sql += ' and ab_itemid is null'
            else:
                sql += ' and ab_itemid = '+m[2]

            cur.execute(sql)
            microevents = cur.fetchall()
            for me in microevents: 
                if(me[0] == None):
                    print 'me date 0'
                    continue;
                wholedata.append(['me',(me[0]-admitdate).total_seconds(),me])

        wholedata = sorted(wholedata, key=itemgetter(1))

        map_time_index = {}
        index = 0;
        for wd in wholedata:
            if(wd[1] not in map_time_index):
                map_time_index[wd[1]] = index
                index += 1;

        # date input output labevent prescript
        D = 2+len(input_itemid)+len(output_itemid)+len(lab_itemid)+len(prescript_itemid)+len(chart_itemid)+len(microbio_itemid)

        patient = [[None for i in range(D)] for j in range(len(map_time_index))]
        numtodivide = [[0 for i in range(D-2)] for j in range(len(map_time_index))]
        print D
        print len(patient)
        
        unitCheckFlag = True
        for wd in wholedata:

            assert patient[ map_time_index[wd[1]] ][D-2] == None or patient[ map_time_index[wd[1]] ][D-2] == wd[1]
            patient[ map_time_index[wd[1]] ][D-2] = wd[1]
            patient[ map_time_index[wd[1]] ][D-1] = aid

            # test units
            if(wd[0] ==  'ie'):
                if( wd[2][3].replace(' ','').lower() == units[map_itemid_index[wd[2][1]]]):
                    pass;
                else:
                    print wd
                    unitCheckFlag = False
                    continue;
            elif(wd[0] ==  'oe'):
                pass;
            elif(wd[0] ==  'pe'):
                if( wd[2][3].replace(' ','').lower() == units[map_itemid_index[wd[2][1]]]):
                    pass;
                else:
                    print wd
                    unitCheckFlag = False
                    continue;
            elif(wd[0] ==  'le'):
                if( wd[2][3].replace(' ','').lower() == units[map_itemid_index[wd[2][1]]]):
                    pass;
                else:
                    print wd
                    unitCheckFlag = False
                    continue;
            elif(wd[0] ==  'ce'):
                pass;

            if(wd[0] ==  'ie' or wd[0] ==  'oe'):
                if(wd[2][2] == None): continue;

                if(patient[map_time_index[wd[1]]][map_itemid_index[wd[2][1]]] ==None):
                    patient[map_time_index[wd[1]]][map_itemid_index[wd[2][1]]] = wd[2][2] 
                else: 
                    patient[map_time_index[wd[1]]][map_itemid_index[wd[2][1]]] += wd[2][2]

            if(wd[0] ==  'pe'):
                dose = wd[2][2]
                dose = dose.replace(',','').replace(' ','')
                numVal = None;
                try:
                    numVal = float(dose)
                except:
                    if(len(dose.split('-')) ==2):
                        strs = dose.split('-')
                        try:
                            numVal = (float(strs[0]) + float(strs[1]))/2.0
                        except:
                            continue;
                    else:
                        continue;

                if(patient[map_time_index[wd[1]]][map_itemid_index[wd[2][1]]] == None):
                    if(numVal != None):
                        patient[map_time_index[wd[1]]][map_itemid_index[wd[2][1]]] = numVal
                else: 
                    if(numVal != None):
                        patient[map_time_index[wd[1]]][map_itemid_index[wd[2][1]]] += numVal

            if(wd[0] ==  'le'):
                dose = wd[2][2]
                if(dose == None): continue;

                if(patient[map_time_index[wd[1]]] [map_itemid_index[wd[2][1]]] == None):
                    patient[  map_time_index[wd[1]]  ][  map_itemid_index[wd[2][1]]  ] = dose
                    numtodivide[map_time_index[wd[1]]  ]  [  map_itemid_index[wd[2][1]] ] = 1
                else: 
                    patient[map_time_index[wd[1]]][map_itemid_index[wd[2][1]]] += dose
                    numtodivide[ map_time_index[wd[1]]  ] [   map_itemid_index[wd[2][1]] ] += 1

            if(wd[0] ==  'ce'):
                dose = wd[2][2]
                if(dose == None): continue;

                if(patient[map_time_index[wd[1]]] [map_itemid_index[wd[2][1]]] == None):
                    patient[  map_time_index[wd[1]]  ][  map_itemid_index[wd[2][1]]  ] = dose
                    numtodivide[map_time_index[wd[1]]  ]  [  map_itemid_index[wd[2][1]] ] = 1
                else: 
                    patient[map_time_index[wd[1]]][map_itemid_index[wd[2][1]]] += dose
                    numtodivide[ map_time_index[wd[1]]  ] [   map_itemid_index[wd[2][1]] ] += 1

            if(wd[0] ==  'me'):
                dose = wd[2][2]
                dose = dose.replace('<','').replace('>','').replace('=','')
                numVal = None;
                if(dose == ''):
                    continue;
                try:
                    numVal = float(dose)
                except:
                    continue;

                if(patient[map_time_index[wd[1]]][map_itemid_index[wd[2][1]]] == None):
                    if(numVal != None):
                        patient[map_time_index[wd[1]]][map_itemid_index[wd[2][1]]] = numVal
                        numtodivide[map_time_index[wd[1]]  ]  [  map_itemid_index[wd[2][1]] ] = 1
                else: 
                    if(numVal != None):
                        patient[map_time_index[wd[1]]][map_itemid_index[wd[2][1]]] += numVal
                        numtodivide[ map_time_index[wd[1]]  ] [   map_itemid_index[wd[2][1]] ] += 1
                        print wd

        for i in range(len(map_time_index)):
            for j in range(D-2):
                if(numtodivide[i][j] == 0): continue;
                patient[i][j] /= numtodivide[i][j]

        if(unitCheckFlag):
            DB.append(patient)
        sys.stdout.flush()
    print "done : "+ str(len(DB))
    return DB
```
