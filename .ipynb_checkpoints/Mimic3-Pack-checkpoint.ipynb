```python

hours_bf_die_st = 48
hours_bf_die_ed = 24
```

```python
period_mortal_mean_perday = [];
period_mortal_std_perday = [];
period_mortal_mean_perapr = [];
period_mortal_std_perapr = [];

counting = []
for f in range(numFeature):
    allDosePerDay = []
    allDosePerApr = []
    for i in range(len(DB_MORTAL)):
        if(ALM_MORTAL[i][5] == 1):
            print 's',
            continue;
        
        sum = 0
        count = 0
        mustLaterThan = (ALM_MORTAL[i][3] - hours_bf_die_ed*60)*60
        mustBeforeThan = (ALM_MORTAL[i][3] - hours_bf_die_st*60)*60
        for j in range(len(DB_MORTAL[i])):
            if(DB_MORTAL[i][j][numFeature] < mustBeforeThan or DB_MORTAL[i][j][numFeature] > mustLaterThan):
                continue;
            if(DB_MORTAL[i][j][f]!=None):
                sum += DB_MORTAL[i][j][f]
                count += 1
        if(count>0):
            allDosePerDay.append(sum / (hours_bf_die_st-hours_bf_die_ed) /60.);
            allDosePerApr.append(sum / count);
    
    period_mortal_mean_perday.append(np.array(allDosePerDay).mean())
    period_mortal_std_perday.append(np.array(allDosePerDay).std())
    period_mortal_mean_perapr.append(np.array(allDosePerApr).mean())
    period_mortal_std_perapr.append(np.array(allDosePerApr).std())
    counting.append(len(allDosePerApr))
    print f,
```
