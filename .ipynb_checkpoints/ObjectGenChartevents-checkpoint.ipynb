```python
# DB connection:
import psycopg2
import datetime
import numpy as np
import sys
try:
    # conn = psycopg2.connect("dbname='mimic' user='kanit' host='melady1.usc.edu' password='abcd6712'")
    conn = psycopg2.connect("dbname='mimic' user='kanit' host='localhost' password='abcd6712'")
except:
    print "I am unable to connect to the database"
print 'connection established'
```

```python
# get admission list (list of hadm_id)
ADMISSION_IDS = []
cur = conn.cursor()
sql = 'SELECT hadm_id FROM mimiciii.ADMISSIONS ORDER BY hadm_id'
cur.execute(sql)
rows  = cur.fetchall()
for row in rows:
    ADMISSION_IDS.append(row[0]);

print 'total admission : '+ str(len(ADMISSION_IDS))
```

```python
# Generate keys
cur = conn.cursor()
cur.execute('select distinct itemid from mimiciii.chartevents')
items = cur.fetchall()

tmp = []
for s in items: tmp.append(s[0])
items=tmp

lenitem = len(items)

map_item = {}

for i in range(len(items)):
    map_item[items[i]] = i
```

```python
PATIENTS_ADM = [];
ADMISSIONS = [];

ICD9 = [];
DRGCODES = [];
PROCEDURES_ICD = [];

CHARTEVENTS = [];

ct=0
# ADMISSION_IDS = [100109]
for aid in ADMISSION_IDS:
#     ct+=1
#     if(ct%100==0):print '.',
    print aid

    # Generate Admission array
    cur = conn.cursor()
    sql = 'SELECT * FROM mimiciii.ADMISSIONS WHERE hadm_id = ' + str(aid)
    cur.execute(sql)
    adm = cur.fetchone()
    ADMISSIONS.append(list(adm))
#     print (admtime-admtime+datetime.timedelta(0,1)).total_seconds()
    
    # Generate Patient array
    cur = conn.cursor()
    sql = 'SELECT * FROM mimiciii.patients WHERE subject_id = ' + str(adm[1])
    cur.execute(sql)
    pat = cur.fetchone()
    PATIENTS_ADM.append(list(pat))
    
    # Get birthdate and admission datetime
    birthdate = pat[3]
    admitdate = adm[3]

    # date | valNum 5561| valTxt 5561 | uomNum 5561| uomTxt 5561 | cgid 5561| cgid 5561
    slot1 = 1
    slot2 = 1 + lenitem
    slot3 = 1 + lenitem + lenitem
    slot4 = 1 + lenitem + lenitem + lenitem
    slot5 = 1 + lenitem + lenitem + lenitem + lenitem
    slot6 = 1 + lenitem + lenitem + lenitem + lenitem + lenitem
    
    cur = conn.cursor()
    cur.execute('SELECT charttime,itemid,value,valuenum,valueuom,cgid FROM mimiciii.chartevents WHERE hadm_id='+str(aid)+' ORDER BY charttime ASC')
    chs = cur.fetchall()
    charttimes = []
    charts = []
    for c in chs:
        if c[0] not in charttimes:
            charttimes.append(c[0])
            charts.append([])
        charts[len(charts)-1].append(c)
    timeseries = [];
        
    # logic = 
    # 1 put first down
    # if duplicated
    # try to merge (numeric + same unit)
    # unreadable -> text field
    # different unit = next row
    for di in range(len(charttimes)):
        d = charttimes[di]
        #timeseries object
        subtimeseries = [(d-admitdate).total_seconds()]+[None]*lenitem+[None]*lenitem+['']*lenitem+['']*lenitem+['']*lenitem+['']*lenitem
        ins = charts[di]
        num_item = [0]*lenitem;
        only_num = [0]*lenitem;
        
        for pres in ins:
            itemid = pres[1]
            val = pres[2]
            valnum = pres[3]
            valuom = pres[4].replace(' ','').lower()
            cgid = pres[5]

            key = map_item[itemid] 
            
            #always placed
            subtimeseries[slot4+key] += val + ' //'
            subtimeseries[slot5+key] += valuom + ' //'
            subtimeseries[slot6+key] += str(cgid) + ' //'

            if(valnum != None):
                if(subtimeseries[slot1+key] == None):
                    subtimeseries[slot1+key] = valnum
                    subtimeseries[slot2+key] = valuom
                    subtimeseries[slot3+key] = str(cgid) +" //"
                    num_item[key]+=1
                elif( subtimeseries[slot2+key] == valuom):
                    subtimeseries[slot1+key] += valnum
                    subtimeseries[slot3+key] += str(cgid) +" //"
                    num_item[key]+=1
        
        for i in range(lenitem):
            if(subtimeseries[slot1+i]!= None):
                subtimeseries[slot1+i] /= num_item[i]
        
        timeseries.append(subtimeseries)
        
    CHARTEVENTS.append(timeseries)
    sys.stdout.flush()
#     break;
    if(len(CHARTEVENTS)==5000):
        np.savez('/home/kanit/mimic3/result/chartevents'+str(ct),CHARTEVENTS)
        ct+=1
        CHARTEVENTS=[]

np.savez('/home/kanit/mimic3/result/chartevents'+str(ct),CHARTEVENTS)
print 'done'    
```
